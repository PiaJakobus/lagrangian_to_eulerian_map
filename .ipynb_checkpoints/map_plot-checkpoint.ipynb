```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import pandas as pd
import itertools
from matplotlib.patches import Circle


data1 = pd.read_csv( 'particles.dat', delim_whitespace=True, header=None)
data_error = pd.read_csv("error.dat", delim_whitespace=True, header=None)


pdim = round(len((np.array(data1)[:,0]))**(1/3)) 
gdim = round(len((np.array(data_error)[:,0]))**(1/3)) 

x1 = np.array(data1.iloc[:, 1]).reshape(pdim,pdim,pdim)
y1 = np.array(data1.iloc[:, 2]).reshape(pdim,pdim,pdim)
z1 = np.array(data1.iloc[:, 3]).reshape(pdim,pdim,pdim)
fp = np.array(data1.iloc[:, 4]).reshape(pdim,pdim,pdim)

r1 = np.array(data_error.iloc[:, 0]).reshape(gdim,gdim,gdim)
r2 = np.array(data_error.iloc[:, 1]).reshape(gdim,gdim,gdim)
r3 = np.array(data_error.iloc[:, 2]).reshape(gdim,gdim,gdim)
err = np.array(data_error.iloc[:, 3]).reshape(gdim,gdim,gdim)


fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
ax.set_aspect('equal')
cb = plt.scatter(y1[:,:,-1],x1[:,:,-1],c=fp[:,:,-1],s=30)
plt.scatter(r1[:,-1,:],r3[:,-1,:],marker="x",color="red",s=100)


plt.colorbar(cb,label=r'$\exp(-r_i^2)$')
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.xlim(-1,1)
plt.ylim(-1,1)
```
